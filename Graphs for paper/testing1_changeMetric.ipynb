## 4. The oscillations in the numerical $S_i$ do not get mcuh smaller when we add the first-order MMR correction, take the previous system as an example

```python
def run_resCorr(sim, correction_term):
    ps = sim.particles
    lsys = LaplaceLagrangeSystem.from_Simulation(sim)
    for Input in correction_term:
        lsys.add_first_order_resonance_term(Input[0], Input[1], Input[2]) # add resonance correction
        
    M = lsys.Neccentricity_matrix
    vals,T = np.linalg.eigh(M) # T returned by eigh is R^T in paper. Could define R = T.T
    
    ec13 = 1-ps[1].a/ps[3].a
    m1, m2, m3 = ps[1].m, ps[2].m, ps[3].m
    m_tot = m1 + m2 + m3
    Tsec = 2*np.pi/(1/2*m_tot/ps[0].m/ec13**2)*ps[1].P

    Nout = 1000
    times = np.linspace(0,2*Tsec, Nout)
    Gx, Gy = np.zeros((Nout,3)), np.zeros((Nout,3))
    for i, time in enumerate(times):
        sim.integrate(time)    
        e1x, e2x, e3x = [p.e*np.cos(p.pomega) for p in ps[1:]]
        e1y, e2y, e3y = [p.e*np.sin(p.pomega) for p in ps[1:]]
        Gx[i] = [np.sqrt(2*p.m*np.sqrt(sim.G*ps[0].m*p.a)*(1-np.sqrt(1-p.e**2)))*np.cos(p.pomega) for p in ps[1:]]  #changed -pomega to pomega
        Gy[i] = [np.sqrt(2*p.m*np.sqrt(sim.G*ps[0].m*p.a)*(1-np.sqrt(1-p.e**2)))*np.sin(p.pomega) for p in ps[1:]]

    Sx = T.T @ Gx.T
    Sy = T.T @ Gy.T
    S1, S2, S3 = np.sqrt(Sx**2 + Sy**2)
   
    return times, S1, S2, S3
```

```python
mu2 = 1/10
mass = (1-mu2)/2*1e-6, mu2*1e-6, (1-mu2)/2*1e-6
ecc = 0.017, 0.012, 0.010
pomega = 0.84, 1.10, -0.73
a13, d = 0.75, -0.2
alpha = solveAlpha(a13,d,'no')[0], solveAlpha(a13,d,'no')[1]
print(alpha)
```

```python
# LL
sim = make_sim(alpha, mass, ecc, pomega)
times, G1, G2, G3, S1, S2, S3, S1a, S2a, S3a, angle1, angle2, angle3, angle1a, angle2a, angle3a = run(sim,"yes")

# LL + correction
sim = make_sim(alpha, mass, ecc, pomega)
times, S1_resCorr, S2_resCorr, S3_resCorr = run_resCorr(sim,[[1,2,6],[2,3,5]])
```
