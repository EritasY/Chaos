```python
import rebound
import numpy as np
import matplotlib.pyplot as plt
from celmech.secular import LaplaceLagrangeSystem
from celmech import Poincare
%matplotlib inline
plt.style.use('paper.mplstyle')
```

```python
sim = rebound.Simulation()
sim.G = 4*np.pi**2
sim.add(m=1)
sim.add(m=1e-8, a=1, e=0.01)
sim.add(m=2e-8, a=1+20*(1e-8/3)**(1/3), e=0.02, f=0.5)
sim.move_to_com()
ps = sim.particles
sim.integrator = "whfast"
sim.dt = 0.05
1-ps[1].a/ps[2].a
```

```python
dpi = 300.4

ecom = 0.1
e12 = 0.15
phicom = np.pi/5
phi12 = np.pi/2.3
ecomx = ecom*np.cos(phicom)
ecomy = ecom*np.sin(phicom)

s=0.95

fig, axarr = plt.subplots(figsize=(16,8), ncols=3)
ax = axarr[0]
e12x = -ps[2].m/(ps[1].m+ps[2].m)*e12*np.cos(phi12)
e12y = -ps[2].m/(ps[1].m+ps[2].m)*e12*np.sin(phi12)

ax.arrow(0, 0, s*ecomx, s*ecomy, head_width=0.01, linewidth=4, head_length=0.02, fc='tab:blue', ec='tab:blue', length_includes_head=True)
circle = plt.Circle((ecomx, ecomy), np.sqrt(e12x**2 + e12y**2), color='tab:orange', fill=False, lw=3)
ax.add_patch(circle)
print(np.sqrt(e12x**2 + e12y**2))

ax.arrow(ecomx, ecomy, s*e12x, s*e12y, head_width=0.01, linewidth=4, head_length=0.02, fc='tab:orange', ec='tab:orange', length_includes_head=True)
ax.arrow(0, 0, s*(ecomx+e12x), s*(ecomy+e12y), head_width=0.01, linewidth=4, head_length=0.02, fc='k', ec='k', length_includes_head=True)


ax = axarr[1]
e12x = ps[1].m/(ps[1].m+ps[2].m)*e12*np.cos(phi12)
e12y = ps[1].m/(ps[1].m+ps[2].m)*e12*np.sin(phi12)

ax.arrow(0, 0, s*ecomx, s*ecomy, head_width=0.01, linewidth=4, head_length=0.02, fc='tab:blue', ec='tab:blue', length_includes_head=True)
circle = plt.Circle((ecomx, ecomy), np.sqrt(e12x**2 + e12y**2), color='tab:orange', fill=False, lw=3)
ax.add_patch(circle)
print(np.sqrt(e12x**2 + e12y**2))

ax.arrow(ecomx, ecomy, s*e12x, s*e12y, head_width=0.01, linewidth=4, head_length=0.02, fc='tab:orange', ec='tab:orange', length_includes_head=True)

ax.arrow(0, 0, s*(ecomx+e12x), s*(ecomy+e12y), head_width=0.01, linewidth=4, head_length=0.02, fc='k', ec='k', length_includes_head=True)

ax = axarr[2]

e12x = e12*np.cos(phi12)
e12y = e12*np.sin(phi12)

circle = plt.Circle((0,0), e12, color='tab:orange', fill=False, lw=3)
ax.add_patch(circle)

ax.arrow(0, 0, s*e12x, s*e12y, head_width=0.01, linewidth=4, head_length=0.02, fc='tab:orange', ec='tab:orange', length_includes_head=True)


axarr[0].text(-0.015, 0.04, r'$e_{com}$', c='tab:blue', fontsize=18)
axarr[0].text(0, -0.05, r'$e_1$', c='k', fontsize=18)
axarr[0].text(0.075, 0.01, r'$-\tilde{m}_2 e_{12}$', c='tab:orange', fontsize=18)

axarr[1].text(0.03, -0.01, r'$e_{com}$', c='tab:blue', fontsize=18)
axarr[1].text(-0.0, 0.06, r'$e_2$', c='k', fontsize=18)
axarr[1].text(0.1, 0.11, r'$\tilde{m}_1 e_{12}$', c='tab:orange', fontsize=18)

axarr[2].text(0.02, 0.03, r'$e_{12}$', c='tab:orange', fontsize=18)

for ax in axarr:
    ax.set_aspect('equal')
    ax.set_xticks([0])
    ax.set_yticks([0])
    ax.set_xlim([-0.2, 0.2])
    ax.set_ylim([-0.2, 0.2])
    ax.set_xticklabels(['0.0'])
    ax.set_yticklabels([])
    ax.grid('on', linestyle='--')

fig.subplots_adjust(hspace=0, wspace=0)
axarr[0].set_yticklabels(['0.0'])

xpos, ypos = -0.18,-0.16
axarr[0].text(xpos, ypos, r'$e_{1}$',horizontalalignment='left', verticalalignment='center', bbox=dict(facecolor='white',edgecolor='k'),fontsize=26)
axarr[1].text(xpos, ypos, r'$e_{2}$',horizontalalignment='left', verticalalignment='center', bbox=dict(facecolor='white',edgecolor='k'),fontsize=26)
axarr[2].text(xpos, ypos, r'$e_{12}$',horizontalalignment='left', verticalalignment='center', bbox=dict(facecolor='white',edgecolor='k'),fontsize=26)
```

```python
fig.savefig('2p.pdf', transparent=True, dpi=dpi, bbox_inches='tight')   
```
