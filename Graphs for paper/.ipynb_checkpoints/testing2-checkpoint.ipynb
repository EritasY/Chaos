```python
def calcS(sim,mode,version):
    ps = sim.particles
    m_tot = ps[1].m + ps[2].m + ps[3].m
    
    m1, m2, m3 = ps[1].m, ps[2].m, ps[3].m
    mu1, mu2, mu3 = m1/m_tot, m2/m_tot, m3/m_tot
    alpha12, alpha23 = ps[1].a/ps[2].a, ps[2].a/ps[3].a
    alpha13 = alpha12*alpha23
    
    ec12 = alpha12**(-1/4)*alpha23**(3/4)*alpha23**(-1/8)*(1-alpha12)
    ec23 = alpha23**(-1/2)*alpha12**(1/8)*(1-alpha23)
    ec13 = alpha13**(-1/2)*(1-alpha13)
        
    Delta = (ec12 - ec23)/ec13
    a = (1-Delta)**3*(3+Delta)
    b = (1+Delta)**3*(3-Delta)
    
    a1, a2, a3 = (ps[1].a), (ps[2].a), (ps[3].a)
    
    if version == "old":
        if mode == 1:
            R1, R2, R3 = np.sqrt(m1*m2), -(m1+m3), np.sqrt(m2*m3)
        elif mode == 2:
            R1, R2, R3 = -np.sqrt(m3), 0, np.sqrt(m1)
        elif mode == 3:
            R1, R2, R3 = np.sqrt(m1), np.sqrt(m2), np.sqrt(m3)
    elif version == "new":
        if mode == 1:
            R1, R2, R3 = np.sqrt(m1*m2)*(a1*a2)**(1/4), -(m1*a1**(1/2)+m3*a3**(1/2)), np.sqrt(m2*m3)*(a2*a3)**(1/4)
        elif mode == 2:
            R1, R2, R3 = -np.sqrt(m3)*(a3)**(1/4), 0, np.sqrt(m1)*(a1)**(1/4)
        elif mode == 3:
            R1, R2, R3 = np.sqrt(m1)*(a1)**(1/4), np.sqrt(m2)*(a2)**(1/4), np.sqrt(m3)*(a3)**(1/4)
    
    c1 = R1*np.sqrt(m1)*(sim.G*ps[0].m*ps[1].a)**(1/4)
    c2 = R2*np.sqrt(m2)*(sim.G*ps[0].m*ps[2].a)**(1/4)
    c3 = R3*np.sqrt(m3)*(sim.G*ps[0].m*ps[3].a)**(1/4)
    return 1/np.sqrt(c1**2+c2**2+c3**2) * np.array([c1, c2, c3])
```

```python
def distS(sim,mode,version):
    ps = sim.particles  
    lsys = LaplaceLagrangeSystem.from_Simulation(sim)  
    M = lsys.Neccentricity_matrix
    vals,T = np.linalg.eigh(M)
    m1, m2, m3 = ps[1].m, ps[2].m, ps[3].m
    
    for i in range(3):   
        T[0][i] = T[0][i]*np.sqrt(m1)*(sim.G*ps[0].m*ps[1].a)**(1/4)
        T[1][i] = T[1][i]*np.sqrt(m2)*(sim.G*ps[0].m*ps[2].a)**(1/4)
        T[2][i] = T[2][i]*np.sqrt(m3)*(sim.G*ps[0].m*ps[3].a)**(1/4)
        
    exp_theo = (T.T)[mode-1]
    vec_theo = 1/np.linalg.norm(exp_theo) * exp_theo
    vec_approx = calcS(sim,mode,version)
    
    dist = min([np.linalg.norm(vec_theo-vec_approx),np.linalg.norm(vec_theo+vec_approx)])
    
    return dist
```

```python
def solveAlpha(a13, Delta, simplified):
    if simplified == 'yes':
        dfunc = lambda a1 : ((1-a1) - (1-(a13/a1))) / ((1-a13)) - Delta
    elif simplified == 'no':
        dfunc = lambda a1 : (a1**(-1/4)*(a13/a1)**(5/8)*(1-a1) - a1**(1/8)*(a13/a1)**(-1/2)*(1-(a13/a1))) / (a13**(-1/2)*(1-a13)) - Delta
    a1_initial_guess = 0.8
    a1_val = fsolve(dfunc, a1_initial_guess)
    return a1_val[0], a13/a1_val[0]

def getT(mass, alpha):    
    m1, m2, m3 = mass
    m_tot = m1 + m2 + m3
    mu1, mu2, mu3 = m1/m_tot, m2/m_tot, m3/m_tot
    alpha12, alpha23 = alpha
    alpha13 = alpha12*alpha23
    
    ec12, ec23, ec13 = 1-alpha12, 1-alpha23, 1-alpha13
    
    Delta = (ec12 - ec23)/ec13
    a = (1-Delta)**3*(3+Delta)
    b = (1+Delta)**3*(3-Delta)
    A = mu1*a/(mu1*a+mu3*b)
    B = mu3*b/(mu1*a+mu3*b)
    
    T = np.array([[A,-(A+B),B],[1,0,-1],[mu1,mu2,mu3]])
    return T

def StoE(mass,alpha,S,phi):
    T = getT(mass,alpha)
    S1, S2, S3 = S
    p1, p2, p3 = phi
    S_vec = np.array([[S1*cmath.exp(1j*p1)], [S2*cmath.exp(1j*p2)], [S3*cmath.exp(1j*p3)]])
    e_vec = np.linalg.inv(T) @ S_vec
    
    e1, e2, e3 = np.abs(e_vec[0][0]), np.abs(e_vec[1][0]), np.abs(e_vec[2][0])
    pom1, pom2, pom3 = cmath.phase(e_vec[0][0]), cmath.phase(e_vec[1][0]), cmath.phase(e_vec[2][0])
    
    return (e1, e2, e3), (pom1, pom2, pom3)

def EtoS(mass,alpha,ecc,pomega):
    e1, e2, e3 = ecc
    pom1, pom2, pom3 = pomega
    
    T = getT(mass,alpha)
    e_vec = np.array([[e1*cmath.exp(1j*pom1)], [e2*cmath.exp(1j*pom2)], [e3*cmath.exp(1j*pom3)]])
    S_vec = T @ e_vec
    
    S1, S2, S3 = np.abs(S_vec[0][0]), np.abs(S_vec[1][0]), np.abs(S_vec[2][0])
    phi1, phi2, phi3 = cmath.phase(S_vec[0][0]), cmath.phase(S_vec[1][0]), cmath.phase(S_vec[2][0])
    
    return (S1, S2, S3), (phi1, phi2, phi3)
```

```python
mass_param = np.array([1e-10, 1/10, 1/5, 1/3])
a13 = 0.78
delta = np.linspace(-0.9,0.9,100)
alpha_param = [[solveAlpha(a13,d,'no')[0], solveAlpha(a13,d,'no')[1]] for d in delta]
ecc = 0.01, 0.01, 0.01
pomega = 0, 2*np.pi/3, -2*np.pi/3
data1, data2 = np.zeros((len(mass_param)*len(alpha_param),4)), np.zeros((len(mass_param)*len(alpha_param),4))

for i, mu2 in enumerate(mass_param):
    mass = (1-mu2)/2*1e-6, mu2*1e-6, (1-mu2)/2*1e-6
    for j, (alpha1, alpha2) in enumerate(alpha_param):
        sim = make_sim((alpha1, alpha2), mass, ecc, pomega)
#         data1[i*len(alpha_param)+j][:] = mu2, distS(sim,1,"old"), distS(sim,2,"old"), distS(sim,3,"old")
#         data2[i*len(alpha_param)+j][:] = mu2, distS(sim,1,"new"), distS(sim,2,"new"), distS(sim,3,"new")
        data1[i*len(alpha_param)+j][:] = mu2, modeDistance(sim, "no", 1), modeDistance(sim, "no", 2), modeDistance(sim, "no", 3)
        data2[i*len(alpha_param)+j][:] = mu2, modeDistance(sim, "yes", 1), modeDistance(sim, "yes", 2), modeDistance(sim, "yes", 3)
df1 = pd.DataFrame(data1, columns=['m2','dist1','dist2','dist3'], index=None)   
df2 = pd.DataFrame(data2, columns=['m2','dist1','dist2','dist3'], index=None) 
```
