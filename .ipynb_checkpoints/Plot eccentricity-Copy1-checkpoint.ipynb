```python
import numpy as np
import matplotlib.pyplot as plt
import rebound
import reboundx
import pandas as pd
from celmech.poincare import Poincare
import scipy.optimize as opt
%matplotlib inline
```

```python
def sim(alpha, mass, ecc, pom, scale):
    alpha12, alpha23 = alpha
    alpha13=alpha12*alpha23
    P1, P2, P3 = alpha12**(3/2), 1, 1/alpha23**(3/2)
    m1, m2, m3 = mass
    m_tot = m1+m2+m3
    ecc1, ecc2, ecc3 = ecc
    pomega1, pomega2, pomega3 = pom
    
    # start simulation
    sim = rebound.Simulation()
    sim.units = ('yr', 'AU', 'Msun')

    # add star, planet 1, planet 2
    sim.add(m=1.)
    sim.add(m=m1, P=P1, e=ecc1, pomega=pomega1)
    sim.add(m=m2, P=P2, e=ecc2, pomega=pomega2)
    sim.add(m=m3, P=P3, e=ecc3, pomega=pomega3)
    ps = sim.particles
    ps[1].r = ps[1].a*(ps[1].m/3/ps[0].m)**(1/3)
    ps[2].r = ps[2].a*(ps[2].m/3/ps[0].m)**(1/3)

    sim.move_to_com()
    sim.integrator = "whfast"
    sim.dt = sim.particles[1].P/20
    sim.collision = "direct"
    
    Nout=400
#     ecom, e13, eminus, ecc1, ecc2, ecc3, S2, S3 = np.zeros(Nout), np.zeros(Nout), np.zeros(Nout), np.zeros(Nout), np.zeros(Nout), np.zeros(Nout), np.zeros(Nout), np.zeros(Nout)
    e13, eminus,S2, S3, K2, K3 = np.zeros(Nout), np.zeros(Nout), np.zeros(Nout), np.zeros(Nout),np.zeros(Nout), np.zeros(Nout)
    times = np.linspace(0,scale,Nout)
    
    r = (1-alpha12)/(1-alpha13)
    C = (2*r-1)/((m1*m2/m_tot-m3)*r**2 + (m2*m3/m_tot-m1)*(1-r)**2 + (m1+m3)**2/m_tot*r**2*(1-r)**2)*m_tot
    w2 = 2*m2/(m1+m3) * (m1/(np.pi*(1-alpha23)**2) + m3/(np.pi*(1-alpha12)**2)) + 2*(m1+m3)/(np.pi*(1-alpha13)**2)
    w3 = 2*m_tot/(m1+m3) * (m1/(np.pi*(1-alpha12)**2) + m3/(np.pi*(1-alpha23)**2))
    k = np.sqrt(m1*m2*m3*m_tot)/(m1+m3) * (2/(np.pi*(1-alpha23)**2) - 2/(np.pi*(1-alpha12)**2))
    phi = 1/2*np.arctan2(2*k, w2-w3)
    
    for i, time in enumerate(times):
        try:
            sim.integrate(time)
            info_1 = sim.particles[1].calculate_orbit(primary=sim.particles[0])
            info_2 = sim.particles[2].calculate_orbit(primary=sim.particles[0])
            info_3 = sim.particles[3].calculate_orbit(primary=sim.particles[0])
            p1, p2, p3, e1, e2, e3 = info_1.pomega, info_2.pomega, info_3.pomega, info_1.e, info_2.e, info_3.e
            e1_x, e1_y = e1*np.cos(p1), e1*np.sin(p1)
            e2_x, e2_y = e2*np.cos(p2), e2*np.sin(p2)
            e3_x, e3_y = e3*np.cos(p3), e3*np.sin(p3)
            
#             ecc1[i] = e1
#             ecc2[i] = e2
#             ecc3[i] = e3
            
#             ecom[i] = np.sqrt((m1*e1_x+m2*e2_x+m3*e3_x)**2 + (m1*e1_y+m2*e2_y+m3*e3_y)**2) / m_tot
            e13[i] = np.sqrt((e3_x-e1_x)**2 + (e3_y-e1_y)**2)
            eminus[i] = np.sqrt((m3*(e3_x-e2_x)-m1*(e2_x-e1_x))**2 + (m3*(e3_y-e2_y)-m1*(e2_y-e1_y))**2) / m_tot
            S2_x = np.cos(phi)*(e3_x-e1_x) + np.sin(phi)*(m3*(e3_x-e2_x)-m1*(e2_x-e1_x))/m_tot
            S2_y = np.cos(phi)*(e3_y-e1_y) + np.sin(phi)*(m3*(e3_y-e2_y)-m1*(e2_y-e1_y))/m_tot
            S3_x = -np.sin(phi)*(e3_x-e1_x) + np.cos(phi)*(m3*(e3_x-e2_x)-m1*(e2_x-e1_x))/m_tot
            S3_y = -np.sin(phi)*(e3_y-e1_y) + np.cos(phi)*(m3*(e3_y-e2_y)-m1*(e2_y-e1_y))/m_tot
            
            S2[i] = np.sqrt(S2_x**2+S2_y**2)
            S3[i] = np.sqrt(S3_x**2+S3_y**2)
            
            K2_x = (e3_x-e1_x) + m2/m_tot*C*(m3*(e3_x-e2_x)-m1*(e2_x-e1_x))/m_tot
            K2_y = (e3_y-e1_y) + m2/m_tot*C*(m3*(e3_y-e2_y)-m1*(e2_y-e1_y))/m_tot
            K3_x = -m1*m3/m_tot**2*C*(e3_x-e1_x) + (m3*(e3_x-e2_x)-m1*(e2_x-e1_x))/m_tot
            K3_y = -m1*m3/m_tot**2*C*(e3_y-e1_y) + (m3*(e3_y-e2_y)-m1*(e2_y-e1_y))/m_tot
            
            K2[i] = np.sqrt(K2_x**2+K2_y**2)
            K3[i] = np.sqrt(K3_x**2+K3_y**2)
                       
        except rebound.Collision:
            break
            

    return times, S2, S3, K2, K3
#     return times, ecc1, ecc2, ecc3, ecom, e13, eminus, S2, S3
```

```python
alpha = 0.89, 0.91
mass = 2e-7,1e-7,1e-7
ecc = 0.02, 0.02, 0.02
pomega = 0, 0, np.pi/3
times, S2, S3, K2, K3 = sim(alpha, mass, ecc, pomega, 2e6)
# times, ecc1, ecc2, ecc3, ecom, e13, eminus, S2, S3 = sim(alpha, mass, ecc, pomega, 2e6)
```

```python
def sim(alpha, mass, ecc, pom, scale):
    alpha12, alpha23 = alpha
    P1, P2, P3 = alpha12**(3/2), 1, 1/alpha23**(3/2)
    mu1, mu3 = mass
    ecc1, ecc2, ecc3 = ecc
    pomega1, pomega2, pomega3 = pom
    
    # start simulation
    sim = rebound.Simulation()
    sim.units = ('yr', 'AU', 'Msun')

    # add star, planet 1, planet 2
    sim.add(m=1.)
    sim.add(m=mu1, P=P1, e=ecc1, pomega=pomega1)
    sim.add(P=P2, e=ecc2, pomega=pomega2)
    sim.add(m=mu3, P=P3, e=ecc3, pomega=pomega3)
    ps = sim.particles
    ps[1].r = ps[1].a*(ps[1].m/3/ps[0].m)**(1/3)
    ps[2].r = ps[2].a*(ps[2].m/3/ps[0].m)**(1/3)

    sim.move_to_com()
    sim.integrator = "whfast"
    sim.dt = sim.particles[1].P/20
    sim.collision = "direct"
    
    Nout=400
    e12, e23 = np.zeros(Nout), np.zeros(Nout)
    times = np.linspace(0,scale,Nout)
    
    for i, time in enumerate(times):
        try:
            sim.integrate(time)
            info_1 = sim.particles[1].calculate_orbit(primary=sim.particles[0])
            info_2 = sim.particles[2].calculate_orbit(primary=sim.particles[0])
            info_3 = sim.particles[3].calculate_orbit(primary=sim.particles[0])
            p1, p2, p3, e1, e2, e3 = info_1.pomega, info_2.pomega, info_3.pomega, info_1.e, info_2.e, info_3.e
            e1_x, e1_y = e1*np.cos(p1), e1*np.sin(p1)
            e2_x, e2_y = e2*np.cos(p2), e2*np.sin(p2)
            e3_x, e3_y = e3*np.cos(p3), e3*np.sin(p3)
            
            e12[i] = np.sqrt((e2_x-e1_x)**2 + (e2_y-e1_y)**2)
            e23[i] = np.sqrt((e3_x-e2_x)**2 + (e3_y-e2_y)**2)
            
        except rebound.Collision:
            break
    return times, e12, e23
```
